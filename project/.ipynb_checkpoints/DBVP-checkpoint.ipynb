# CAPS
Cargamos los datos crudos desde el excel, descartamos las columnas innecesarias y cambiamos el nombre a las columnas según la necesidad. Además creamos dos tablas, una para VP y otra para VU.

```python
qcap_path = "../AUDITORIAS/CAPS/1-  PUNTO QCAP.xlsm"
caps = pd.read_excel(qcap_path, "DATOS")
caps.drop(columns=['MES', 'SEM', 'Caja de velocidades', 'Fabrica', 'Titulo defecto', 'Comentario defecto', 'Elementos', 'Unnamed: 21', 'Unnamed: 22', 'Unnamed: 23', 'Unnamed: 24','Unnamed: 25','Unnamed: 26', 'Unnamed: 27', 'Unnamed: 28', 'Unnamed: 29', 'Taller'], inplace=True)
caps[['INC','ELE']] = caps['Codigo defecto'].str.split("-",expand=True)
caps.rename(columns={'Familia':'FLIA', 'Usina':'USINA','Tipo  vehiculo':'TIPO', 'Avance producto':'DPTO', 'Motor':'MOTOR', 'Trasmision':'TRANS', 'Version':'VERS', 'Num. de Identificacion':'PJI', 'Codigo defecto':'DEF', 'Demerito (pt)':'DEM'}, inplace=True)
# caps.head()
caps['AUDI'] = 'CAPS'
caps = caps[['AUDI','FECHA', 'USINA', 'FLIA', 'TIPO', 'PJI', 'ELE', 'INC', 'DEM', 'DPTO']]
caps.replace({'DPTO': 'TCP'}, 'PINTURA', inplace=True)
caps.replace({'DPTO': 'TCT'}, 'SOLDADURA', inplace=True)
caps = caps[caps['PJI'].notna()]
caps['PJI'] = caps['PJI'].apply(lambda x: str(int(x)))
```

# CAPS VP

```python
filtvp = (caps['USINA']=='VP')
caps_vp = caps[filtvp]
caps_vp_generic = caps_vp[['AUDI', 'FECHA', 'FLIA', 'TIPO', 'PJI', 'ELE', 'INC', 'DEM', 'DPTO']]
```

# PESD VP
Cargamos los datos crudos desde el excel, descartamos las columnas innecesarias y cambiamos el nombre a las columnas.

```python
pesd_vp_path = "../AUDITORIAS/PESD/VP/6-PESD VP - copia.xlsm"
pesd_vp = pd.read_excel(pesd_vp_path, "Datos crudos")
pesd_vp.drop(columns=['Mes', 'Code site', 'Point de déclaration', 'Semana2','A','B','C','D','E', 'F', 'DEFECTO','Incidente2', 'Elemento2', 'FAMILIA','DEFECTO MODELO', 'Defecto2', 'Semana', 'Mes2'], inplace=True)
pesd_vp.rename(columns={'Code Localisation':'LOC', 'DEPARTAMENTO':'DPTO', 'COD DEFECTO':'DEF','Incidente':'INC', 'Elemento':'ELE','Date de déclaration':'FECHA', 'Démérite':'DEM', 'Famille':'FLIA','Type':'TIPO', 'PJI avec défaut':'PJI', 'Code UET imputée':'UET', 'Code regroupement de pilotage':'TIPODEF'}, inplace=True)
pesd_vp.dropna(subset=['DEF'], how='any', inplace=True)
# pesd_vp = pesd_vp[pesd_vp['ELE'].notna()]
# pesd_vp = pesd_vp[pesd_vp['INC'].notna()]
pesd_vp['AUDI'] = 'PESD'
pesd_vp['ELE'] = pesd_vp['ELE'].astype(int)
pesd_vp['INC'] = pesd_vp['INC'].astype(int)
pesd_vp['PJI'] = pesd_vp['PJI'].apply(lambda x: str(int(x)))
```

```python
pesd_vp_generic = pesd_vp[['AUDI', 'FECHA', 'FLIA', 'TIPO', 'PJI', 'ELE', 'INC', 'LOC', 'DEM', 'TIPODEF', 'DPTO', 'UET']]
pesd_vp_generic.head()
```

```python
pesd_vp_generic['DPTO'].replace(regex=r'^MONTAJE.*', value='MONTAJE', inplace=True)
pesd_vp_generic['DPTO'].replace(regex=r'^EMBUTI.*', value='EMBUTICION', inplace=True)
```

```python
saves_vp_path = "../AUDITORIAS/SAVES/VP/8-SAVES VP.xlsm"
saves_vp = pd.read_excel(saves_vp_path, "Datos")
saves_vp.drop(columns=['Año', 'Mes', 'GEOMETRIA', 'OPERADOR FUGA CHECKMAN', 'ZONA', 'PILOTO', 'DETECCION PINTURA', 'DETECCION MONTAJE', 'RETOCADOR PINTURA', 'RETOCADOR MONTAJE', 'Z', 'RETOCADOR MONTAJE','PESD', 'PUESTO FUGA CHECKMAN', 'PUESTO FUGA CSC /CSCD', 'ZONA CHECKMAN', 'OPERADOR FUGA CSC/CSCD', 'Semana', 'Fabrica', 'NA AR', 'PRO-PIE', 'DEFECTO', 'Defecto', 'ORIGEN', 'VALIDADOR','Retocador','UET'], inplace=True)
saves_vp[['INC','ELE']] = saves_vp['Codigo'].str.split("-",expand=True)
saves_vp.rename(columns={'Avance PIEZA':'PIEZA', 'Codigo':'DEF', 'Numero de fabricacion':'NFAB', 'Día':'FECHA', 'FAMILIA':'FLIA', 'Tipo  vehiculo':'TIPO', 'Motor':'MOTOR', 'Caja de velocidades':'CVELO', 'Trasmision':'TRANS', 'Version':'VERS', 'Num. de Identificacion':'PJI', 'DEMERITO':'DEM', 'GRUPOS':'TIPODEF'}, inplace=True)
# saves_vp['Avance PIEZA'].value_counts(dropna=False)
saves_vp['AUDI'] = 'SAVES'
# saves_vp.head(5)
```

```python
saves_vp_generic = saves_vp[['AUDI', 'FECHA', 'FLIA', 'TIPO', 'PJI', 'ELE', 'INC', 'DEM', 'TIPODEF', 'DPTO']]
saves_vp_generic['TIPODEF'].replace({'ASPECTO': 'ASP', 'GEOMETRIA':'GMTR', 'MODO OPERATORIO':'MOP', 'FUNCIONAMIENTO':'FCIO', 'RUIDOS':'RDOS', 'DEGRADACION':'DGRC', 'FALTANTE':'FTES', 'NO CONFORME':'NCON'}, inplace=True)
saves_vp_generic.head()
```

# DSTAR VP

```python
import datetime
my_date = datetime.date.today() # if date is 01/01/2018
year, my_week, day_of_week = my_date.isocalendar()
# print("Week #" + str(my_week-1) + " of year " + str(year))
dstar_path = "../AUDITORIAS/DSTAR/VP/DSTAR KO VP S" +  str(my_week-1) + ".xlsx"
dstar = pd.read_excel(dstar_path, "Defectos")
new_header = dstar.iloc[0]
dstar = dstar[1:]
dstar.columns = new_header
dstar.rename(columns={'Fórmula PJI':'PJI', 'Famille':'FLIA', 'Code incident':'INC', 'Code élément':'ELE', 'Démérite':'DEM', 'Code Localisation':'LOC', 'Horodatage de déclaration':'FECHA', 'Code UET imputée':'UET', 'Code département imputé':'DPTO','Horodatage de retouche':'FECHA RET'}, inplace=True)
dstar['AUDI'] = 'DSTAR'
# dpu_vp.rename(columns={'fecha':'FECHA', 'Famille':'FLIA', 'Dép.':'DPTO', 'loc':'LOC', 'Type':'TIPO', 'cod elem':'ELE', 'cod inc':'INC', 'dem':'DEM', 'Dép':'DPTO','Ate.':'TALLER', 'Code regroupement de pilotage':'TIPODEF'}, inplace=True)
```

```python
# dstar.head()
# FALTA ELE, TALLER
# FALTA VU
dstar_generic = dstar[['AUDI', 'FECHA', 'PJI', 'FLIA', 'INC', 'ELE', 'DEM', 'LOC', 'DPTO', 'UET']]
dstar_generic['DPTO'].replace({'DCHAP':'SOLDADURA', 'DPINT':'PINTURA', 'DMONT':'MONTAJE', 'EMB':'EMBUTICION'}, inplace=True)
```

```python
dstar_generic.head()
```

```python
table = pd.concat([caps_vp_generic, pesd_vp_generic, saves_vp_generic, dpu_vp_generic, dstar_generic], sort=True)
table = pd.merge(table, elementos, on='ELE', how='left')
table = pd.merge(table, incidentes, on='INC', how='left')
```

```python
table['DEFECTO'] = table['ELEMENTO'] + ' ' + table['INCIDENTE']
table.to_excel("datos_vp.xlsx") 
```
